```python
#import
import pandas as pd
```

```python
# set path to CSV files
school_csvpath = "Instructions/PyCitySchools/Resources/schools_complete.csv"
student_csvpath = "Instructions/PyCitySchools/Resources/students_complete.csv"
```

```python
school_df = pd.read_csv(school_csvpath)
student_df = pd.read_csv(student_csvpath)
```

```python
#Merge data into one df
merged_df = pd.merge(school_df, student_df, on="school_name", how="left")
merged_df.head()
```

```python
#Calculate Number of schools
number_schools = merged_df["school_name"].nunique()
number_schools
```

```python
# Calculate the number of students
number_students = merged_df["Student ID"].nunique()
number_students
```

```python
# Calculate total budget
total_budget = school_df["budget"].sum()
total_budget
```

```python
# Calculate the avg. math score
avg_math = merged_df["math_score"].mean()
avg_math
```

```python
# Calc av reading score
avg_reading = merged_df["reading_score"].mean()
avg_reading
```

```python
# determine number students with >=70 math score
pass_math = (merged_df["math_score"]>= 70).sum()
# Calc percentage students with math score 70 or higher
per_pass_math = (pass_math/number_students)*100
per_pass_math.round(6)
```

```python
# determine number students with >=70 math score
pass_reading = (merged_df["reading_score"]>= 70).sum()
# Calc percentage students with math score 70 or higher
per_pass_reading = (pass_reading/number_students)*100
per_pass_reading.round(6)
```

```python
# Calc number of students with reading & math scores 70 or higher

pass_both_df = merged_df[(merged_df["reading_score"] >=70) & (merged_df["math_score"] >=70)]
pass_StID = pass_both_df["Student ID"].count()  

# Calc percentage studens 
per_pass_both = (pass_StID/number_students)*100
per_pass_both.round(6)

 
```
