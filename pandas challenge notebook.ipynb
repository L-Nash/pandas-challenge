```python
#import
import pandas as pd
```

```python
# set path to CSV files
school_csvpath = "Instructions/PyCitySchools/Resources/schools_complete.csv"
student_csvpath = "Instructions/PyCitySchools/Resources/students_complete.csv"
```

```python
csv_school = pd.read_csv(school_csvpath)
student_df = pd.read_csv(student_csvpath)
```

```python
#Merge data into one df
merged_df = pd.merge(csv_school, student_df, on="school_name", how="left")
merged_df.head()
```

```python
#Calculate Number of schools
number_schools = merged_df["school_name"].nunique()
number_schools
```

```python
# Calculate the number of students
number_students = merged_df["Student ID"].nunique()
number_students
```

```python
# Calculate total budget
total_budget = csv_school["budget"].sum()
total_budget
```

```python
# Calculate the avg. math score
avg_math = merged_df["math_score"].mean()
avg_math
```

```python
# Calc av reading score
avg_reading = merged_df["reading_score"].mean()
avg_reading
```

```python
# determine number students with >=70 math score
pass_math = (merged_df["math_score"]>= 70).sum()
# Calc percentage students with math score 70 or higher
per_pass_math = (pass_math/number_students)*100
per_pass_math.round(6)
```

```python
# determine number students with >=70 math score
pass_reading = (merged_df["reading_score"]>= 70).sum()
# Calc percentage students with math score 70 or higher
per_pass_reading = (pass_reading/number_students)*100
per_pass_reading.round(6)
```

```python
# Calc number of students with reading & math scores 70 or higher

pass_both_df = merged_df[(merged_df["reading_score"] >=70) & (merged_df["math_score"] >=70)]
pass_StID = pass_both_df["Student ID"].count()  

# Calc percentage studens 
per_pass_both = (pass_StID/number_students)*100
per_pass_both.round(6)

 
```

```python
# Create dataframe to hold the results (see starter .py for fomatting)

summary_df = pd.DataFrame({"Number of Schools": [number_schools], 
                           "Number of Students": [number_students], 
                          "Total Budget": [total_budget],
                           "Average Math Score": [avg_math], 
                           "Average Reading Score": [avg_reading], 
                          "% Passing Math": [per_pass_math], 
                            "% Passing Math": [per_pass_reading], 
                            "Overall Passing": [per_pass_both]})

# Format Results
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("{:.1f}%".format)
summary_df["Overall Passing"] = summary_df["Overall Passing"].map("{:.1f}%".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:.1f}%".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:.1f}%".format)

summary_df
```

```python
# Create overview table per school, to include:    
#   * School Name
#   * School Type
#   * Total Students
#   * Total School Budget
#   * Per Student Budget
#   * Average Math Score
#   * Average Reading Score
#   * % Passing Math
#   * % Passing Reading
#   * % Overall Passing (The percentage of students that passed math **and** reading.)


School_df = merged_df.groupby(["school_name"])
```

```python
#   * School Name
school_name_sh = School_df["school_name"].first()
```

```python
#   * School Type

school_type_sh = School_df["type"].first()
```

```python
#   * Total Students
total_student_sh = School_df["type"].count()
```

```python
#   * Total School Budget
total_budget_sh = School_df["budget"].first()
```

```python
#   * Per Student Budget
per_stu_budget_sh =  total_budget_sh / total_student_sh
```

```python
#   * Average Math Score
avg_math_sh = School_df["math_score"].mean()
```

```python
#   * Average Reading Score
avg_reading_sh = School_df["reading_score"].mean()
```

```python
# find stud passing reading
passing_reading = merged_df[merged_df["reading_score"]>=70]
pass_reading_stu = passing_reading ["school_name"].value_counts()
per_pass_reading_sh = (pass_reading_stu / total_student_sh)*100

```

```python
# find students passing math
passing_math = merged_df[merged_df["math_score"]>=70]
pass_math_stu = passing_math["school_name"].value_counts()
per_pass_passing_math_sh = (pass_math_stu  / total_student_sh)*100


```

```python
# combine
passing_both_sh = merged_df[(merged_df["math_score"]>=70) & (merged_df["reading_score"]>=70)]
```

```python
# Calc percentage of students passing overall
pass_both_stu = passing_both_sh ["school_name"].value_counts()
per_overall_pass_sh = (pass_both_stu / total_student_sh)*100
```

```python
# Create a dataframe for overview per school results

school_summary_df = pd.DataFrame({"School Type": school_type_sh, 
                "Total Students": total_student_sh, 
                "Total School Budget": total_budget_sh,
                "Per Student Budget": per_stu_budget_sh, 
                "Average Math Score": avg_math_sh,
                "Average Reading Score": avg_reading_sh,
                "% Passing Math": per_pass_passing_math_sh, 
                "% Passing Reading": per_pass_reading_sh, 
                "% Overall Passing": per_overall_pass_sh})

re_school_summary_df = school_summary_df.rename(columns={"school_name":"School Name "})

#format
re_school_summary_df["Total School Budget"] = re_school_summary_df["Total School Budget"].map("${:,.2f}".format)
re_school_summary_df["Per Student Budget"] = re_school_summary_df["Per Student Budget"].map("${:,.2f}".format)
re_school_summary_df["% Passing Math"] = re_school_summary_df["% Passing Math"] .map("{:.1f}%".format)
re_school_summary_df["% Passing Reading"] = re_school_summary_df["% Passing Reading"].map("{:.1f}%".format)
re_school_summary_df["% Overall Passing"] = re_school_summary_df["% Overall Passing"].map("{:.1f}%".format)
re_school_summary_df["Average Math Score"] = re_school_summary_df["Average Math Score"].map("{:.1f}%".format)
re_school_summary_df["Average Reading Score"] = re_school_summary_df["Average Reading Score"].map("{:.1f}%".format)
re_school_summary_df
```

```python
# * Sort and display the top five performing schools by % overall passing.
Top_sh = school_summary_df.sort_values(["% Overall Passing"],ascending=False)
Top_sh.head(5)

Top_sh["Total School Budget"] = Top_sh["Total School Budget"].map("${:,.2f}".format)
Top_sh["Per Student Budget"] = Top_sh["Per Student Budget"].map("${:,.2f}".format)
Top_sh["% Passing Math"] = Top_sh["% Passing Math"] .map("{:.1f}%".format)
Top_sh["% Passing Reading"] = Top_sh["% Passing Reading"].map("{:.1f}%".format)
Top_sh["% Overall Passing"] = Top_sh["% Overall Passing"].map("{:.1f}%".format)
Top_sh["Average Math Score"] = Top_sh["Average Math Score"].map("{:.1f}%".format)
Top_sh["Average Reading Score"] = Top_sh["Average Reading Score"].map("{:.1f}%".format)
Top_sh.head(5)
```

```python
# * Sort and display the five worst-performing schools by % overall passing.
Worst_sh = school_summary_df.sort_values(["% Overall Passing"],ascending=True)
 
Worst_sh["Total School Budget"] = Worst_sh["Total School Budget"].map("${:,.2f}".format)
Worst_sh["Per Student Budget"] = Worst_sh["Per Student Budget"].map("${:,.2f}".format)
Worst_sh["% Passing Math"] = Worst_sh["% Passing Math"] .map("{:.1f}%".format)
Worst_sh["% Passing Reading"] = Worst_sh["% Passing Reading"].map("{:.1f}%".format)
Worst_sh["% Overall Passing"] = Worst_sh["% Overall Passing"].map("{:.1f}%".format)
Worst_sh["Average Math Score"] = Worst_sh["Average Math Score"].map("{:.1f}%".format)
Worst_sh["Average Reading Score"] = Worst_sh["Average Reading Score"].map("{:.1f}%".format)
Worst_sh.head(5)
```

```python
# Create four series that hold the average math score per grade for each school
nine_grade_math = merged_df[(merged_df["grade"] == "9th")].groupby(["school_name"]).mean()["math_score"]
ten_grade_math = merged_df[(merged_df["grade"] == "10th")].groupby(["school_name"]).mean()["math_score"]
eleven_grade_math = merged_df[(merged_df["grade"] == "11th")].groupby(["school_name"]).mean()["math_score"]
twelve_grade_math = merged_df[(merged_df["grade"] == "12th")].groupby(["school_name"]).mean()["math_score"]
```

```python
# combine scores into a dataframe
math_score_df = pd.DataFrame({"9th": nine_grade_math, 
                                "10th": ten_grade_math,
                                 "11th": eleven_grade_math, 
                                 "12th": twelve_grade_math})

math_score_df ["9th"] = math_score_df ["9th"].map("{:.1f}%".format)
math_score_df["10th"] = math_score_df ["10th"].map("{:.1f}%".format)
math_score_df["11th"] = math_score_df ["11th"].map("{:.1f}%".format)
math_score_df["12th"] = math_score_df ["12th"].map("{:.1f}%".format)
math_score_df
```

```python
# Create four series that hold the average reading score per grade for each school
nine_grade_reading = merged_df[(merged_df["grade"] == "9th")].groupby(["school_name"]).mean()["reading_score"]
ten_grade_reading = merged_df[(merged_df["grade"] == "10th")].groupby(["school_name"]).mean()["reading_score"]
eleven_grade_reading = merged_df[(merged_df["grade"] == "11th")].groupby(["school_name"]).mean()["reading_score"]
twelve_grade_reading = merged_df[(merged_df["grade"] == "12th")].groupby(["school_name"]).mean()["reading_score"]
 
```

```python
# combine reading scores into a dataframe
reading_score_df = pd.DataFrame({"9th": nine_grade_reading, 
                                "10th": ten_grade_reading,
                                 "11th": eleven_grade_reading, 
                                 "12th": twelve_grade_reading})

reading_score_df ["9th"] = reading_score_df["9th"].map("{:.1f}%".format)
reading_score_df["10th"] = reading_score_df["10th"].map("{:.1f}%".format)
reading_score_df["11th"] = reading_score_df["11th"].map("{:.1f}%".format)
reading_score_df["12th"] = reading_score_df["12th"].map("{:.1f}%".format)
reading_score_df
```

```python
# * Create a table that breaks down school performances based on average Spending Ranges (Per Student). 

bins = [0, 585, 630, 645, 680]
cost = ["<$585", "$585-$630", "$630-$645", "$645-$680"]


school_summary_df["Per Student Budget Summary"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=cost, include_lowest=True)

Per_stu_budget = school_summary_df.groupby(["Per Student Budget Summary"]).mean()

Per_stu_budget["% Passing Math"] = Per_stu_budget["% Passing Math"] .map("{:.1f}%".format)
Per_stu_budget["% Passing Reading"] = Per_stu_budget["% Passing Reading"].map("{:.1f}%".format)
Per_stu_budget["% Overall Passing"] = Per_stu_budget["% Overall Passing"].map("{:.1f}%".format)
Per_stu_budget["Average Math Score"] = Per_stu_budget["Average Math Score"].map("{:.1f}%".format)
Per_stu_budget["Average Reading Score"] = Per_stu_budget["Average Reading Score"].map("{:.1f}%".format)


Per_stu_budget.iloc[:, 3:8] 
 
```

```python
# * Perform the same operations as above, based on school size.

bins = [0, 1000, 2000, 5000]
size = ["< 1000", "1000-2000", "2000-5000"]

school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins, labels=size, include_lowest=True)

Per_sh_size = school_summary_df.groupby(["School Size"]).mean()



Per_sh_size["% Passing Math"] = Per_sh_size["% Passing Math"] .map("{:.1f}%".format)
Per_sh_size["% Passing Reading"] = Per_sh_size["% Passing Reading"].map("{:.1f}%".format)
Per_sh_size["% Overall Passing"] = Per_sh_size["% Overall Passing"].map("{:.1f}%".format)
Per_sh_size["Average Math Score"] = Per_sh_size["Average Math Score"].map("{:.1f}%".format)
Per_sh_size["Average Reading Score"] = Per_sh_size["Average Reading Score"].map("{:.1f}%".format)
Per_sh_size.iloc[:, 3:8]

 
```
